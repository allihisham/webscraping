```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# Set up Selenium
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
service = Service('C:/webdrivers/chromedriver.exe')  # Update with your path to chromedriver
driver = webdriver.Chrome(service=service, options=chrome_options)

# URL of the page to scrape
url = 'https://www.4g.co.uk/sim-only/'  # Replace with the actual URL
driver.get(url)

# Function to click the "Show More" button
def click_show_more():
    try:
        show_more_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.ID, 'show-more-button'))
        )
        # Scroll the button into view
        driver.execute_script("arguments[0].scrollIntoView();", show_more_button)
        time.sleep(1)  # Wait for the scroll to complete
        show_more_button.click()
        time.sleep(2)  # Wait for the new content to load
    except Exception as e:
        print(f"Error clicking 'Show More' button: {e}")

# Click "Show More" button multiple times
click_count = 3
for _ in range(click_count):
    click_show_more()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all deals in the HTML
deals = soup.find_all('div', class_=lambda x: x and ('sim deal' in x or 'sim complianceContainer deal' in x))

# Extract and print details for each deal
for deal in deals:
    title = deal.find('span', class_='name').text
    network_span = deal.find('span', class_='mvno')
    network = network_span.text if network_span else "N/A"
    data = deal.find('div', class_='data box').find('p', class_='value').text
    price = deal.find('div', class_='pricing box').find('p', class_='value').text
    link = deal.find('a', class_='dealLink')['href']
    
    # Extract additional details
    details = deal.find('div', class_='details').find_all('p')
    contract_length = details[0].text.strip() if len(details) > 0 else "N/A"
    minutes = details[1].text.strip() if len(details) > 1 else "N/A"
    texts = details[2].text.strip() if len(details) > 2 else "N/A"
    
    # Extract gift and benefit
    gift = deal.find('div', class_='btnCustom contentModalBtn gift')
    gift_text = gift.find('span', class_='innerText').text if gift else "N/A"
    benefit = deal.find('div', class_='btnCustom contentModalBtn benefit')
    benefit_text = benefit.find('span', class_='innerText').text if benefit else "N/A"
    
    # Extract popular status
    popular = deal.find('div', class_='popular')
    popular_text = popular.text.strip() if popular else "N/A"
    
    print(f"Title: {title}")
    print(f"Network: {network}")
    print(f"Data: {data}")
    print(f"Price: {price}")
    print(f"Contract Length: {contract_length}")
    print(f"Minutes: {minutes}")
    print(f"Texts: {texts}")
    print(f"Gift: {gift_text}")
    print(f"Benefit: {benefit_text}")
    print(f"Popular: {popular_text}")
    print(f"Link: {link}")
    print("-" * 40)

# Close the browser
driver.quit()
```
